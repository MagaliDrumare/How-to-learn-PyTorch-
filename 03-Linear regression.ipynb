## Linear Regression 
Simple linear regression basics 
* y=Ax+b 
* y=2x+1 

## Aim of the linear regression : 
Minimize the distance between the point and the line 
calculate the distance through MSE 
calculate the gradients 
update the parameters A and b 
parameters = parameters-learning rate*gradients 


## Build a linear regression model on CPU and GPU 
Step 1 : create the model class 
Step 2 : instantiate the model class 
Step 3 : Instantiate the Loss Class 
Step 4: Instantiate the Optimizer Class 
Step 5: Train the model 

## Things to be on GPU : Variables and models 
model_name.cuda()
variable_name.cuda()


```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np

# create the dataset 
x_values = [i for i in range(11)]
x_train = np.array(x_values, dtype=np.float32)
x_train = x_train.reshape(-1, 1)

y_values = [2*i + 1 for i in x_values]
y_train = np.array(y_values, dtype=np.float32)
y_train = y_train.reshape(-1, 1)

'''
Create the class 
'''
class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)  
    
    def forward(self, x):
        out = self.linear(x)
        return out

'''
instantiate the model 
'''
input_dim = 1
output_dim = 1

model = LinearRegressionModel(input_dim, output_dim)


#######################
#  USE GPU FOR MODEL  #
#######################



'''
instantiate the loss
'''

criterion = nn.MSELoss()

'''
instantiate the optimizer
'''

learning_rate = 0.01

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

'''
train the models 
'''
epochs = 100
for epoch in range(epochs):
    epoch += 1
    # Convert numpy array to torch Variable
    
    #######################
    #  USE GPU FOR MODEL  #
    #######################
    #if torch.cuda.is_available():
    inputs = Variable(torch.from_numpy(x_train))
        
    #######################
    #  USE GPU FOR MODEL  #
    #######################
    # torch.cuda.is_available():
    labels = Variable(torch.from_numpy(y_train))
        
    # Clear gradients w.r.t. parameters
    optimizer.zero_grad() 
    
    # Forward to get output
    outputs = model(inputs)
    
    # Calculate Loss
    loss = criterion(outputs, labels)
    
    # Getting gradients w.r.t. parameters
    loss.backward()
    
    # Updating parameters
    optimizer.step()
    
    # Logging
    print('epoch {}, loss {}'.format(epoch, loss.data[0]))

```
